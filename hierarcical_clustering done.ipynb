```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
dataset = pd.read_csv('rent_lagos_updated.csv')
dataset
```

```python
dataset = dataset.drop(columns = ['housing_type', 'bathrooms', 'guest_toilet', 'parking_space', 'district', 'address', 'local government'])
```

```python
dataset = dataset.replace(['ketu ojota'], '32')
dataset
```

```python
n = 10000
newDataSet = dataset.iloc[:n] 
newDataSet
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
for x in newDataSet.index:
    if newDataSet.loc[x, "bedrooms"] == 6:
        newDataSet.drop(x, inplace = True)
        
newDataSet
```

```python
for x in newDataSet.index:
    if newDataSet.loc[x, "bedrooms"] == 7:
        newDataSet.drop(x, inplace = True)
        
newDataSet
```

```python
for x in newDataSet.index:
    if newDataSet.loc[x, "bedrooms"] == 9:
        newDataSet.drop(x, inplace = True)
        
newDataSet
```

```python
for x in newDataSet.index:
    if newDataSet.loc[x, "bedrooms"] == 5:
        newDataSet.drop(x, inplace = True)
        
newDataSet
```

```python
newDataSet = newDataSet.drop(columns = ['bedrooms'])
```

```python
plt.figure(figsize=(10, 6))
plt.title("House rent prices in Lagos Dendogram")
dendrogram = sch.dendrogram(sch.linkage(newDataSet, method = 'ward'))
```

```python
plt.scatter(newDataSet['districtID'], newDataSet['rent_per_annum'])
```

```python
#performing ah clustering
hc = AgglomerativeClustering(n_clusters = 4, metric = 'euclidean', linkage = 'ward')
```

```python
y_hc = hc.fit_predict(newDataSet)
```

```python
import numpy as np
```

```python
newDataSet = np.array(newDataSet)
plt.scatter(newDataSet[y_hc == 0,1], newDataSet[y_hc == 0, 0], s = 20, c = 'cyan')
plt.scatter(newDataSet[y_hc == 1,1], newDataSet[y_hc == 1, 0], s = 20, c = 'green')
plt.scatter(newDataSet[y_hc == 2,1], newDataSet[y_hc == 2, 0], s = 20, c = 'red')
plt.scatter(newDataSet[y_hc == 3,1], newDataSet[y_hc == 3, 0], s = 20, c = 'yellow')

plt.title("Scatter Plots of House rents in lagos")
plt.ylabel("Rent per annum / 10 million naira")
plt.xlabel("District ID")

#cluster 1, cyan = 1 bedroom
#cluster 2, green = 2 bedroom
#cluster 3, red = 3 bedroom
#cluster 4, yello4 4 bedroom
```
